```python
import polars as pl 
import plotly.express as px
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df =  (
    pl.read_csv(
        r"C:\Users\faune\numerical-finance\Numerical_Finance\NumericalFinance\Results\european_kakutani_antithetic.csv"
    )
     .rename({'meanPrice': 'mean_price', ' nbSimul': 'nb_sim'})
     .with_columns(enter_interval = pl.col(' var').mul(6146.559999999999))
)
```

```python
df.row(-1) #price, var, sims, sims for ci
```

```python


price = 3.7143

sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
sns.lineplot(data=df.select('mean_price', 'nb_sim').to_pandas().set_index('nb_sim'))

plt.axhspan(price - 0.025*price,price + 0.025*price, color='gray', alpha=0.2, label='Target CI')

plt.axhline(y=price, color='black', linestyle='--', linewidth=1.2, label=f'price = {price}')

plt.title("Kakutani and antithetic variance reduction")
plt.legend()
```
